```python
# Building the model
m = 0
c = 0
m_iter = []
c_iter = []
y_pred_iter = []

L = 0.0001  # The learning Rate
epochs = 15000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m =  sum((-2/n) * (X * (Y - Y_pred)))  # Derivative wrt m
    D_c = sum((-2/n) *(Y - Y_pred))  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    m_iter.append(m)
    c_iter.append(c)
    y_pred_iter.append(Y_pred)
print("m: ", m, "c: ", c)
```

```python
loss = []
for i in range(len(m_iter)):
    loss.append(1/(len(Y)) * (sum(Y - y_pred_iter[i])) ** 2)
plt.scatter(m_iter, loss, c='blue', label='Scatter Plot')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.legend()
plt.show()
```
