```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Library untuk melakukan ploting 3D
%matplotlib notebook
```

```python
X = np.array([0.5, 2.3, 2.9])
Y = np.array([1.4, 1.9, 3.2])
```

```python
plt.scatter(X, Y, c='blue', label='Scatter Plot')

plt.xlabel("Weight")
plt.ylabel("Height")
plt.legend()
plt.show()
```

Dari data di atas kita akan coba melakukan fiting garis, diketahu rumus:

Prediksi Y = intercept + slope x Weigth

Pertama kita akan mencari intercept terlebih dahulu, lalu jika kita sudam memahami cara kerja gradient descent kita akan mencari intercept dan slope. Untuk sekarang kita akan memasukan nilai slope sebesar 0.64.
Pertama kita akan memilih angka random untuk intercept dimana angka ini nantinya akan kita optimasi dengan gradient descent.

```python
#menentukan intercept dan slope
temp_intercept = []
intercept = 0
slope = 0.64

# menentukan nilai min max
max_x = np.max(X)
min_x = np.min(X)

#melakukan kalkulasi fiting garis
x = np.linspace(min_x, max_x, 3)
y = intercept + slope * x

#ploting line
plt.plot(x, y, color='red', label='Garis Regresi')

#ploting scatter
plt.scatter(X, Y, c='blue', label='Plot x,y')

plt.xlabel("Weight")
plt.ylabel("Height")
plt.legend()
plt.show()
```

Setelah kita menentukan nilai slope dan intercept, kita akan menghitung seberapa baguskan data dengan garis yang kita buat menggunakan metode Sum of the squred residual.
Perlu di ketahui, nilai residual dikatakan baik jika  mendekati 0 (Nol).

```python
temp_residual = []
sum_of_the_residual = 0

for i in range(len(x)):
    sum_of_the_residual += (Y[i] - (intercept + slope * X[i])) ** 2

print("Nilai Residual jika intercepet 0 : ",sum_of_the_residual)
temp_residual.append(sum_of_the_residual)
```

Nilai residual ini sendiri nantinya kita dapat ploting ke dalam graphing yang berbeda

```python
#melakukan kalkulasi fiting garis
x = np.linspace(min_x, max_x, 3)
y = intercept + slope * x

#nilai intercepet yg kita ubah di masukan ke dalam temp intercept
temp_intercept.append(intercept)

#ploting line
plt.plot(x, y, color='red', label='Garis Regresi')

#ploting scatter
plt.scatter(X, Y, c='blue', label='Plot x,y')

plt.xlabel("Weight")
plt.ylabel("Height")
plt.legend()
plt.show()

#ploting nilai residual
plt.scatter(temp_intercept, temp_residual, c='red', label='Residual')

plt.xlabel("Intercepet")
plt.ylabel("Residual")
plt.legend()
plt.show()
```

# Menggunakan Gradient Descent

```python
intercept = [0]
new_intercept = 0
slope = 0
```

```python
for i in range(99):
    slope += -2 * (Y[0] - (new_intercept + 0.64 * X[0]))
    slope += -2 * (Y[1] - (new_intercept + 0.64 * X[1]))
    slope += -2 * (Y[2] - (new_intercept + 0.64 * X[2]))
    step_size = slope * 0.1
    new_intercept = new_intercept-step_size
    intercept.append(new_intercept)
    slope = 0
```

```python
#melakukan kalkulasi fiting garis
x = np.linspace(min_x, max_x, 3)
#ploting line
for i in range(10):
    plt.plot(x, intercept[i] + 0.64 * x, color='red')

#ploting scatter
plt.scatter(X, Y, c='blue', label='Plot x,y')

plt.xlabel("Weight")
plt.ylabel("Height")
plt.legend()
```

```python
#ploting nilai residual
temp_residual = []
for i in range(100):
    sum_of_the_residual = 0
    for j in range(len(x)):
        sum_of_the_residual += (Y[j] - (intercept[i] + 0.64 * X[j])) ** 2
    temp_residual.append(sum_of_the_residual)
    sum_of_the_residual = 0
print(temp_residual[-1])
    
plt.scatter(intercept, temp_residual, c='red', label='Residual')

plt.xlabel("Intercepet")
plt.ylabel("Residual")
plt.legend()
plt.show()
```

```python
slope_iter_value = [] #nantinya di gunakan untuk melakukan ploting untuk setiap iterasi, sementara kita simpan di dalam list
intercept_iter_value = [] #nantinya di gunakan untuk melakukan ploting untuk setiap iterasi, sementara kita simpan di dalam list
n_intercept = 0 #nilai ini nantinya akan di ubah seiiring berjalannya iterasi
n_slope = 1 #nilai ini nantinya akan di ubah seriing dengan berjalannya iterasi
d_intercept = 0
d_slope = 0
learning_rate = 0.01
for i in range(1000):
    for j in range(len(X)):
        d_intercept += -2 * (Y[j] - (n_intercept + n_slope * X[j]))
        d_slope += -2 * X[j] * (Y[j] - (n_intercept + n_slope * X[j]))
    
    n_intercept = n_intercept - (d_intercept * learning_rate)
    n_slope = n_slope - (d_slope * learning_rate)
    slope_iter_value.append(n_slope)
    intercept_iter_value.append(n_intercept)
    
    d_intercept = 0 # kita kembalikan nilai turunan intercept menjadi 0 agar bisa di kalkulasikan lagi
    d_slope = 0 #kita kembali nilai turunan slope menjadi 0 agar bisa di kelakulasikan lagi
```

```python
# menentukan nilai min max
max_x = np.max(X)
min_x = np.min(X)
plt.figure(figsize=(10,10))
#melakukan kalkulasi fiting garis
x = np.linspace(min_x, max_x, 3)
#ploting line
for i in range(0, len(intercept_iter_value), 100):
    plt.plot(x, intercept_iter_value[i] + slope_iter_value[i] * x, color='red')

#ploting scatter
plt.scatter(X, Y, c='blue', label='Plot x,y')

plt.xlabel("Weight")
plt.ylabel("Height")
plt.legend()
```

```python
residual_iter_value = []
sum_of_the_residual = 0

for i in range(1000):
    for j in range(len(X)):
        sum_of_the_residual += (Y[j] - (intercept_iter_value[i] + slope_iter_value[i] * X[j])) ** 2
    residual_iter_value.append(sum_of_the_residual)
    sum_of_the_residual = 0
print(residual_iter_value[-1])
```

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(slope_iter_value, residual_iter_value, intercept_iter_value, c='r', marker='o')

ax.set_xlabel('Slope')
ax.set_ylabel('Resiual')
ax.set_zlabel('Intercept')

plt.show()
```
